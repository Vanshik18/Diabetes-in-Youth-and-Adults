```python

# Create a new column for age groups
diabetic_data['Age_Group'] = pd.cut(diabetic_data['Age'], bins=bins, labels=labels, right=False)

# Count the number of diabetic patients in each age group
age_group_counts = diabetic_data['Age_Group'].value_counts().sort_index()

# Plotting
plt.figure(figsize=(8, 6))
sns.barplot(x=age_group_counts.index, y=age_group_counts.values, palette='viridis')

# Add labels and title
plt.title('Number of Diabetic Patients by Age Group', fontsize=16)
plt.xlabel('Age Group', fontsize=14)
plt.ylabel('Number of Patients', fontsize=14)
plt.xticks(fontsize=12)
```

```python
# Separate data for Type 1 and Type 2 diabetes
type1_data = diabetic_data[diabetic_data['Diabetes_Type'] == 'Type 1']
type2_data = diabetic_data[diabetic_data['Diabetes_Type'] == 'Type 2']

# Count the number of patients in each age group for Type 1 and Type 2 diabetes
type1_counts = type1_data['Age_Group'].value_counts().sort_index()
type2_counts = type2_data['Age_Group'].value_counts().sort_index()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=type1_counts.index, y=type1_counts.values, color='blue', alpha=0.7, label='Type 1 Diabetes')
sns.barplot(x=type2_counts.index, y=type2_counts.values, color='orange', alpha=0.7, label='Type 2 Diabetes')

# Add labels and title
plt.title('Number of Diabetic Patients by Age Group and Diabetes Type', fontsize=14)
plt.xlabel('Age Group', fontsize=14)
plt.ylabel('Number of Patients', fontsize=10)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Diabetes Type', fontsize=8)
plt.tight_layout()
plt.show()
```
